# Ряд Тэйлора для вычисления экспоненты

$$ e^x = \sum_{k=0}^\infty {x^k \over k!}$$

```python
import math

ITERATIONS = 40


def factorial(n):  # вычисление факториала числа
    factorial = 1
    for i in range(2, n + 1):
        factorial *= i
    return factorial


def my_exp(x):  # ряд Тэйлора (экспонента)
    partial_sum = 0
    for i in range(ITERATIONS):
        partial_sum += (x ** i) / factorial(i)
    return partial_sum


print(math.exp(10))  # встроенный
print(my_exp(10))  # мой
```

# График (накладываются друг на друга, ♥все точно♥)

```python
from IPython.display import set_matplotlib_formats

set_matplotlib_formats('pdf', 'svg')

import matplotlib.pyplot as plt
import numpy as np

v_exp = np.vectorize(my_exp)
print(my_exp, v_exp)

x = np.r_[-10:10:0.001]  # Область определения (для всех x)

plt.plot(x, np.exp(x), label='библиотечный $e^x$')
plt.plot(x, v_exp(x), label='самодельный $e^x$')
plt.xlabel('Ось абсцисс')
plt.ylabel('Ось ординат')
plt.legend()
plt.show()
```
